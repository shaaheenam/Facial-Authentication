# Getting the images

```python
TrainingImagePath='/content/drive/My Drive/Colab Notebooks/CV/Assg3/Dataset/Final Training Images'
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    shear_range = 0.1,
    zoom_range = 0.1,
    horizontal_flip = True
)
test_datagen = ImageDataGenerator()
```

```python
training_set = train_datagen.flow_from_directory(
    TrainingImagePath,
    target_size = (64,64),
    batch_size = 20,
    class_mode = 'categorical'
)
```

```python
test_set = test_datagen.flow_from_directory(
    TrainingImagePath,
    target_size = (64,64),
    batch_size = 20,
    class_mode = 'categorical'
)
```

```python
test_set.class_indices
```

# Creating the CNN face recognition model

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPool2D
from keras.layers import Flatten
from keras.layers import Dense
```

*   2 convolution layers
*   2 max pooling
*   1 flattening
*   1 ANN
*   Output

```python
cnn = Sequential()
cnn.add(Convolution2D(32, kernel_size=(5,5), strides=(1,1), input_shape=(64,64,3), activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
cnn.add(Convolution2D(64, kernel_size=(5,5), strides=(1,1), activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
cnn.add(Flatten())
cnn.add(Dense(64, activation='relu'))
cnn.add(Dense(2, activation='softmax'))
```

```python
cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cnn.fit_generator(
    training_set,
    #steps_per_epoch=20,
    epochs=20,
    validation_data=test_set,
    validation_steps=10
)
```

```python
cnn.save('face_rec.h5')
#import keras.models as models
#var = models.load_model(face_rec.h5)
```
