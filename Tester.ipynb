```python
from google.colab import drive
drive.mount('/content/drive')
import os
os.chdir('/content/drive/My Drive/Colab Notebooks/CV/Assg3/Dataset')
base_dir = os.getcwd()
print(base_dir)
```

```python
import keras.models as models
vari = models.load_model('/content/drive/My Drive/Colab Notebooks/CV/Assg3/face_rec.h5')
```

```python
import numpy as np
from keras.preprocessing import image
#%pylab inline
import matplotlib.pyplot as plt
```

```python
for file in os.listdir(base_dir + '/Final Testing Images'):
  file_name, file_extension = os.path.splitext(file)
  if (file_extension in ['.png','.jpg','.jpeg']):
    ImagePath=base_dir + '/Final Testing Images/' + str(file)
    #print(ImagePath)
    test_image=image.img_to_array(image.load_img(ImagePath, target_size=(64,64)))
    test_image=np.expand_dims(test_image,axis=0)

    result=vari.predict(test_image,verbose=0)
    #print(result[0][0])
    plt.imshow(image.load_img(ImagePath, target_size=(64,64)))
    plt.show()  
    if result[0][0] > result[0][1]: 
      print('No')
    else:
      print('Yes')
    print('prediction is: ',result)
```
