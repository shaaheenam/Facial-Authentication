```python
base_dir = os.getcwd()
print(base_dir)
prototxt_path = os.path.join(base_dir + '/model_data/deploy.prototxt')
caffemodel_path = os.path.join(base_dir + '/model_data/weights.caffemodel')
```

```python
model = cv2.dnn.readNetFromCaffe(prototxt_path, caffemodel_path)
```

```python
count = 0
for file in os.listdir(base_dir + '/Testing Images Raw'):
	file_name, file_extension = os.path.splitext(file)
	if (file_extension in ['.png','.jpg','.jpeg']):
		image = cv2.imread(base_dir + '/Testing Images Raw/' + file)
		print(file)

		(h, w) = image.shape[:2]
		blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))

		model.setInput(blob)
		detections = model.forward()

		# Identify each face
		for i in range(0, detections.shape[2]):
			box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
			(startX, startY, endX, endY) = box.astype("int")

			confidence = detections[0, 0, i, 2]

			# If confidence > 0.5, save it as a separate file
			if (confidence > 0.5):
				count += 1
				frame = image[startY:endY, startX:endX]
				cv2.imwrite(base_dir + '/Dataset/Final Testing Images/' + 'Image' +str(count) + '.jpg', frame)
				print('image written: '+str(count))
		#print('in outer for loop'+str(file))

print("Extracted " + str(count) + " faces from all images")
```
